```python
import pandas as pd
import numpy as np
from jenkspy import jenks_breaks

# Load CSV
data = pd.read_csv(r"F:\50,6.csv")

# Assume the numeric column is named 'Value'
values = data['Value'].values

# Calculate 5-class Jenks breaks
breaks = jenks_breaks(values, n_classes=5)
print("Breakpoints:", breaks)

# Classify each value
def classify(value, breaks):
    for i in range(1, len(breaks)):
        if value <= breaks[i]:
            return i
data['Jenks_Class'] = data['Value'].apply(lambda x: classify(x, breaks))

# Save to CSV
data.to_csv(r"F:\80classified.csv", index=False)
print("Done! Classification saved.")
```

# Summarize Exposure Index

```python
import pandas as pd
import numpy as np

# Load the data file
input_file = r"F:\table.csv"  # â† change this to your actual file path
df = pd.read_csv(input_file)

# Check column names
print("Columns:", df.columns)

# Define the 6 exposure index columns
exposure_cols = [
    'exposure_index_7',
    'exposure_index_8',
    'exposure_index_9',
    'exposure_index_10',
    'exposure_index_11',
    'exposure_index_12'
]
# Replace these with your actual column names

# Build the aggregation dictionary
agg_dict = {}
for col in exposure_cols:
    agg_dict[f"{col}_sum"] = (col, 'sum')
    agg_dict[f"{col}_count"] = (col, lambda x: x.notna().sum())

# Group by country and calculate the sums and counts
result = df.groupby('Main River Basin', dropna=False).agg(**agg_dict).reset_index()

# Export the result to a CSV file
output_file = r"F:\HPI_country_summary2.csv"
result.to_csv(output_file, index=False)

print("Summary completed! Results saved to:", output_file)
print(result.head())
```

# Summarize Risk Classification

```python
import pandas as pd

# Load your dataset
df = pd.read_csv(r"F:\table.csv")

# Check the unique risk categories (optional, just to make sure they are correct)
print(df['Risk Classification'].unique())

# Group by Country and Risk Classification, then count how many stations
risk_counts = df.groupby(['Country', 'Risk Classification']).size().reset_index(name='Count')

# Pivot table to make it more readable (countries as rows, risk classes as columns)
risk_pivot = risk_counts.pivot(index='Country', columns='Risk Classification', values='Count').fillna(0)

# Optional: sort columns in a fixed order if needed
order = ['Not exposed', 'Very Low', 'Low', 'Moderate', 'High', 'Very High']
risk_pivot = risk_pivot.reindex(columns=order)

# Display result
print(risk_pivot)

# Save to Csv
risk_pivot.to_excel(r"F:\risk_summary_by_country.csv")
```
